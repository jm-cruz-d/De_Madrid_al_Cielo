```python
def resizeDataset(path):
    folders = {}
    for pc in glob.glob(f"./{path}/*/*"):
        building = pc.split('/')[-2] 
        folders[building]=[]
    for pc in glob.glob(f"./{path}/*/*"):
        building = pc.split('/')[-2] 
        shape = cv2.imread(pc).shape
        if shape[1]/shape[0] > 1:
            im = cv2.resize(cv2.imread(pc), (480,360))
            cropp1 = im[:360, :360]
            cropp2 = im[:360, -360:]
            folders[building].append(cropp1)
            folders[building].append(cropp2)
        else:
            im = cv2.resize(cv2.imread(pc), (360,480))
            cropp1 = im[:360, :360]
            cropp2 = im[-360:, :360]
            folders[building].append(cropp1)
            folders[building].append(cropp2)
    X = np.concatenate([folders[k] for k,v in folders.items()])
    return X
```
